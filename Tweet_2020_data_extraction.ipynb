```python
# Creating list to append tweet data to
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('covid vaccine until:2020-04-16 since:2020-03-16').get_items()):
    if i>500000:
        break
    tweets_list1.append([tweet.date,tweet.user.username,tweet.content,tweet.lang,tweet.hashtags,tweet.inReplyToUser,
                       tweet.likeCount,tweet.retweetCount,tweet.replyCount])
    
# Creating a dataframe from the tweets list above
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Date','User','Tweet','Language','Hashtags','In_reply','Likes','Retweets','Replies'])
```

```python
tweets_df1.to_csv('First Extract.csv')
```
